```python
sns.scatterplot(x=df['wifi'],y=df['clock_speed'])
plt.title("Distibution of wifi and clock speed")
plt.show()
```

```python
plt.figure(figsize=(12,12))
sns.scatterplot(x=df['px_width'],y=df['px_height'],hue=df['wifi'],palette='summer')
plt.title("Distibution of width and height")
plt.show()
```

```python
sns.histplot(df['dual_sim'],kde=True)
plt.title("Dual sim distribution")
plt.show()
```

```python
sns.countplot(x=df['three_g'],color='orange')
plt.title('Distribution of Three G')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x=df['int_memory'],y=df['clock_speed'],hue=df['n_cores'],palette='coolwarm')
plt.title("Distribution of clock speed")
plt.show()
```

```python
df_filt = df[df['pc']>250]
sns.histplot(df_filt['pc'],kde=True)
plt.show()
```

```python
df_filt = df[df['ram']>250]
sns.histplot(df_filt['ram'],kde=True)
plt.show()
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(df.corr(),cmap='summer',annot=True)
plt.title("Correlation heatmap")
plt.show()
```
